```
from sklearn import model_selection, datasets, linear_model, metrics

import numpy as np
import pandas as pd
```

```
iris = datasets.load_iris()
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(iris.data, iris.target, 
                                                                                     test_size = 0.25)

```

```
classifier = linear_model.SGDClassifier()
classifier.get_params().keys()
```

```
parameters_grid = {
    'loss' : ['hinge', 'log', 'squared_loss'],
    'penalty' : ['l1', 'l2'],
    'max_iter' : np.arange(1, 10),
    'alpha' : np.linspace(0.0001, 0.001, 5),
}
```

```
cv1 = model_selection.StratifiedKFold(n_splits=10)
cv2= model_selection.StratifiedShuffleSplit(n_splits=10, test_size=0.25)
```

```
grid_cv = model_selection.GridSearchCV(classifier, parameters_grid, scoring = 'accuracy', cv = cv1)
```

```
%%time
grid_cv.fit(train_data, train_labels)
```

```

metrics.accuracy_score(grid_cv.best_estimator_.predict(test_data), test_labels)
```

```
print(grid_cv.best_score_)
print(grid_cv.best_params_)
```
