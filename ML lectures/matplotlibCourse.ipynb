```
import numpy as np

import matplotlib.pyplot as plt
```

```
# Create a figure containing a single axes.
plt.grid()
plt.plot(np.array([1, 4, 2, 3])) 
```

```
x = np.linspace(0, 2, 100)
plt.figure(figsize=(7, 7)) 

plt.plot(x, x, label=r'$y=x$')  
plt.plot(x, x**2, label=r'$y=x^2$')  
plt.plot(x, x**3, label=r'$y=x^3$', color='red')
plt.xlabel('x values', size=15)
plt.ylabel('y values', size=15)
plt.title('x by y', size=15)
plt.legend(loc='best')

plt.show()



```

```
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter(x='a', y='b', c='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1, figsize=(8, 8))
ax1 = plt.subplot(121)
ax1.set_xlabel('x values graph 1')
ax1.set_ylabel('y values graph 1')
ax1.set_title('some graph')
ax1.plot(t1, f(t1))
ax2 = plt.subplot(122)
ax2.plot(np.arange(len(t1)), f(t1))



plt.figure(2, figsize=(4, 4))

plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, bins=50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, s=r'$\mu=100\, \sigma=15$', fontsize=20, color='red')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
# link to the math expr https://matplotlib.org/tutorials/text/mathtext.html
```

```


t = np.arange(0, 5, 0.01)
s = t**2 - 2*t + 4
plt.plot(t, s)

plt.annotate('global min', xy=(t[np.argmin(s)], min(s)), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.),
             )

plt.ylim(-0.1, 9)
plt.xlim(-0.1, 5)
plt.show()
```

```
x = np.linspace(-1, 1, 1000)
sine = np.sin(x)
exponent = np.exp(x)
cosine = np.cos(x)
plt.plot(x, sine, label='sin x')
plt.plot(x, exponent, label='exp x')
plt.plot(x, cosine, label='cos x')
plt.axis([-0.5, 1, -1.05, 2])
plt.legend(loc='best')
```

```
from mpl_toolkits import mplot3d

x = np.outer(np.linspace(-5, 5, 30), np.ones(30))
y = x.copy().T
z = np.sin(x + y)

plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('3D plot')
ax.set_xlabel('random values between 0 and 1')
ax.set_ylabel('also random values')
plt.show()
```

```
x = np.linspace(-1, 1, 30)
y = np.sin(x)
ax = plt.axes(projection='3d')
z = np.array(x**2 + y**2)
ax.scatter(xs=x, ys=y, zs=z, zdir='z', s=40, c=np.arange(len(x)))
plt.show()
```

```
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi*2:20j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(x, y, z)
plt.show()
```
