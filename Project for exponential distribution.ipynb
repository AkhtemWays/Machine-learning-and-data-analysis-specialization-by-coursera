```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as sts
%matplotlib inline
```

```python
lambd = 2 # Берем параметр, равный 2 экспоненциального распределения и находим мат. ожидание и дисперсию
E = 1./lambd   #выборочное среднее по определению экспоненциального определения(мат. ожидание)
D = 1./lambd**2 # Дисперсия случайной величины данного распределения
E, D
```

```python
expon_rv = sts.expon(lambd) # Создаем случайную экспоненциальную величину
sample = expon_rv.rvs(1000) # генерация выборки объема 1000
x = np.linspace(expon_rv.ppf(0.01), expon_rv.ppf(0.99),100) # задание интервала для плотности распределения с.в.
pdf = expon_rv.pdf(x) # определение плотности с.в. на интервале
plt.plot(x,pdf, label='Theoretical PDF')
plt.hist(sample, bins = 80, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
```

Оценивание распределения выборочного среднего случайной величины при изменении объёмов выборок



n = 5

```python
n = 5
sample_5 = np.array([])
i = 1
while i <= 1000:
    sample_5i = expon_rv.rvs(n) #генерирование выборки
    mean_5i = sum(sample_5i)/n # вычисление выборочного среднего
    sample_5 = np.append(sample_5, mean_5i) # добавление в массив нового значения
    i+=1
# Определеяем параметры для нормального распределения при n = 5
E = 0.5 # Математическое ожидание для с.в., принадлежащей экспоненциальному распределению
D_5 = D/n # Дисперсия для нормального распределения, приближающая выборку средних
sigma_5 = math.sqrt(D_5) # параметр сигма
norm_5 = sts.norm(E, sigma_5)
pdf_5 = norm_5.pdf(x[:100])
plt.plot(x[:100], pdf_5, label='PDF with n = 5')
plt.hist(sample_5, bins = 80, normed=True)
plt.xlabel('$x values$')
plt.ylabel('$f(x) for n = 5$')
plt.legend(loc='best')
```
